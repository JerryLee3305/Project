### 데이터 불러오기 전처리 과정 

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import matplotlib
```

```python
cust_df = pd.read_csv(r"C:\Users\zzzss\OneDrive\바탕 화면\비타민\11주차\train.csv")
#shape과 head이용해서 데이터 확인하기
cust_df.shape
cust_df.head()
```

```python
#Target data 데이터 갯수 확인
cust_df['TARGET'].value_counts()
```

```python
#데이터 전체 성격 파악
cust_df.describe()
```

```python
# var3열의 min값 -999999 -> 2 값 대체
cust_df['var3'].replace(-999999,2)
#ID 열 삭제
cust_df.drop('ID',axis = 1, inplace = True)
```

```python
# 피처 세트와 레이블 세트분리. 레이블 컬럼은 DataFrame의 맨 마지막에 위치해 컬럼 위치 -1로 분리
# 데이터 개수가 지나치게 많기에 시간내 효율성 위해 10,000개의 데이터만 slicing
X = cust_df.drop('TARGET', axis =1)
y = cust_df['TARGET']
y = y[:10000]
X = X[:10000]
print('피처 데이터 shape:{0}'.format(X.shape))
```

## XGBClassifier 

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.2, random_state=0)
train_cnt = y_train.count()
test_cnt = y_test.count()
print('학습 세트 Shape:{0}, 테스트 세트 Shape:{1}'.format(X_train.shape , X_test.shape))
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score

# n_estimator = 500, random = 156으로 모델 선언. 
xgb_clf =XGBClassifier(n_estimator = 500, random = 156)

# 성능 평가 지표를 auc로 사용, early_stopping_rounds = 100 설정.. 
xgb_clf.fit(X_train, y_train, eval_metric = 'auc',early_stopping_rounds = 100, eval_set=[(X_train, y_train), (X_test, y_test)])

xgb_roc_score = roc_auc_score(y_test, xgb_clf.predict_proba(X_test)[:,1],average='macro')
print('ROC AUC: {0:.4f}'.format(xgb_roc_score))
```

```python
from sklearn.model_selection import GridSearchCV

# n_estimators = 100으로 모델 다시 선언
xgb_clf = XGBClassifier(n_estimators = 100, random = 156)

#parameter = {'max_depth':[5, 7] , 'min_child_weight':[1,3] ,'colsample_bytree':[0.5, 0.75] } 지정해주기
params = {'max_depth':[5,7], 'min_child_weight' : [1,3], 'colsample_bytree':[0.5,0.75]}

# cv = 3, 그리고서치 모델 학습
gridcv = GridSearchCV(xgb_clf, param_grid=params, cv=3)
#early = 30, eval_metric = 'auc', eval_set도 위의 내용과 동일하게 설정.
gridcv.fit(X_train, y_train, early_stopping_rounds=30, eval_metric="auc",
           eval_set=[(X_train, y_train), (X_test, y_test)])
```

```python
#best parameter
print('GridSearchCV 최적 파라미터:',gridcv.best_params_) 
```

```python
xgb_roc_score = roc_auc_score(y_test, gridcv.predict_proba(X_test)[:,1], average='macro')
print('ROC AUC: {0:.4f}'.format(xgb_roc_score))
```

```python
# n_estimator = 500, learning_rate = 0.02, reg_alpha = 0.03로 파라미터 추가하고, best_param 기준 따라 새롭게 파라미터 재정비
xgb_clf = XGBClassifier(n_estimator = 500, learning_rate = 0.02, reg_alpha = 0.03)

# evaluation metric = 'auc', early stopping = 200 후 학습 
xgb_clf.fit(X_train, y_train, eval_metric = 'auc', early_stopping_rounds = 200,eval_set=[(X_train, y_train), (X_test, y_test)])

xgb_roc_score = roc_auc_score(y_test, xgb_clf.predict_proba(X_test)[:,1],average='macro')
print('ROC AUC: {0:.4f}'.format(xgb_roc_score))
```

```python
#가장 중요한 feature 확인하기
from xgboost import plot_importance
import matplotlib.pyplot as plt
%matplotlib inline

fig, ax = plt.subplots(1,1,figsize=(10,8))
plot_importance(xgb_clf, ax=ax , max_num_features=20,height=0.4)
```
